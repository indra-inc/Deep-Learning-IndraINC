## 16.1 Text Mining

#### 16.1.1 Understanding Text Data

```python
## Determining number of words in document
from string import punctuation
from collections import Counter
```

```python
## Displaying the Punctuation marks from string module
print('Different Punctuation marks in string module are :','\n',punctuation)
```

```python
## Reading the document
textfile = open(r"D:\Datasets\textminingwiki.txt",'r')
textdoc = textfile.read()
```

```python
## Printing the document
print('-------------------Original Document of Text Mining-----------------------')
print('')
print(textdoc)
```

```python
## Use of translate() function to remove punctuation marks in document
## Use of lower() function to convert text in lowercase
## Use of split() function to split the document in words

word_doc = textdoc.translate(punctuation).lower().split()
print('-------------------Converting document into lower case words-----------------------')
print('')
print(word_doc)
print('')
print('The number of words in the document is :',len(word_doc))
```

```python
## Use of Counter()function to display the most common words
word_counts = Counter(word_doc)
print('The most common words in document are :','\n')
for word in word_counts.most_common(11):
    print(word)
```

## 16.1.2 Text Preprocessing

```python
## Program related to data preprocessing
## Importing necessary libraries
import nltk
import re
```

```python
## Determining number of words in the original document
print('Number of words in the original document is :',len(textdoc))
```

```python
## Removing unwanted characters from the document using regular expressions
textdoc = re.sub(r'[^a-zA-Z0-9\s]','',textdoc,re.I|re.A)

## Determining number of words in the user created document
print('Number of words after removing unwanted characters',len(textdoc))
```

```python
## Converting all the words to lower case for exact determination
textdoc = textdoc.lower()

## Removing leading and trailing spaces from document using strip()
textdoc = textdoc.strip()

## Displaying the document
print('The new document is :','\n',textdoc)
```

## 16.1.3 Shallow Parsing

```python
## Tokenize the document into tokens/words using word_tokenize() function
nltk.download('punkt')
doc_tokens = nltk.word_tokenize(textdoc)
print('The tokens are :','\n',doc_tokens)
```

```python
## Displaying the nature of the word
nltk.download('averaged_perceptron_tagger')
nltk_pos_tagged = nltk.pos_tag(doc_tokens)

## Displaying tokens (words) and the POS tag
print('The words are :','\n',nltk_pos_tagged)
```

## 16.1.4 Stop Words

```python
## Downloading stop words from nltk library
nltk.download('stopwords')

## Storing all stop words of English from nltk corpus
eng_stopwords = nltk.corpus.stopwords.words('english')

## Determining the number of English stop words in corpus
print('Number of English stop words in corpus are :',len(eng_stopwords))
```

```python
## Displaying the first 30 stop words
print('Example of first 30 stop words include :','\n',eng_stopwords[0:30])
```

```python
## Filtering the tokens which are not in the list of stop words
final_tokens = [token for token in doc_tokens
                           if token not in eng_stopwords]

## Joining the filtered tokens using join() function
newdoc = ' '.join(final_tokens)
```

```python
## Determining the number of words after removal of stopwords in document
print('Number of words after removal of stop words :',len(newdoc))
print('')
print('First 300 letters from my document are :','\n',newdoc[0:300])
```

## 16.1.5 Stemming and Lemmatizing

```python
## Importing necessary libraries
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

## Download the file with the command
nltk.download('wordnet')
ps = PorterStemmer()
lem = WordNetLemmatizer()
```

```python
## Creating a list of words having same base word
my_words = ['program','programs','programming','programer','programers']
for word in my_words:
    print('Stemming word ',word,':',ps.stem(word))
    print('Lemmatizing word ',word,':',lem.lemmatize(word,'v'))
```

```python
## Use of stemming and lemmatization on data
lemlist = []
pslist = []

## Creating a list of words having same base word
for word in final_tokens:
    lemlist.append(lem.lemmatize(word,'v'))
    pslist.append(ps.stem(word))
```

```python
## Displaying the result after lemmatization and stemming
print('After Lemmatization :','\n',lemlist[0:20])
print('After Stemming :','\n',pslist[0:20])
```

## 16.1.6 Word Cloud

```python
## Importing necessary libraries for creating visual display data
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Creating a word cloud of words without lemmatization
start_all_words = ' '
stopwords = set(STOPWORDS)

for word in doc_tokens:
    start_all_words = start_all_words + word + ' '
start_wordcloud = WordCloud(width = 1000, height = 1000,
                           background_color = 'white',stopwords = stopwords,
                           min_font_size = 10).generate(start_all_words)
```

```python
## Creating a word cloud of words with lemmatization
clean_all_words = ' '

for newword in lemlist:
    clean_all_words = clean_all_words + newword + ' '
clean_wordcloud = WordCloud(width = 1000, height = 1000,
                           background_color = 'black',stopwords = stopwords,
                           min_font_size = 10).generate(clean_all_words)
```

```python
## Plot both the word cloud images
plt.figure(1, figsize = (20,10))
plt.subplot(121)
plt.imshow(start_wordcloud)
plt.subplot(122)
plt.imshow(clean_wordcloud)
plt.axis('off')
plt.tight_layout(pad = 0)
plt.show()
```
