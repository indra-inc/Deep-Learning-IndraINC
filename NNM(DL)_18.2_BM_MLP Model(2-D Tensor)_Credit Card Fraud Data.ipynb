```python
# Split into input (X) and output (y) variables
X = credit_card.iloc[:,0:30]
y = credit_card.iloc[:,30]
```

```python
model_1.summary()
```

```python
## Step- 3 : Compile the model
model_1.compile(loss = 'binary_crossentropy', optimizer = 'SGD', metrics = ['accuracy'])
```

```python
## Step- 4 : Fit the model
model_1.fit(x_train, y_train, epochs = 5, batch_size = 1000)
```

```python
## Step- 5 : Evaluate the model
score = model_1.evaluate(x_train, y_train)
# print('Training Accuracy is : %.2f%%' %(100*score[1]))
```

```python
## Step- 5 : Evaluate the model
score_trn = model_1.evaluate(x_train, y_train)
print('Training Accuracy is : %.2f%%' %(100*score_trn[1]))
print('')

score_tst = model_1.evaluate(x_test, y_test)
print('Test Accuracy is : %.2f%%' %(100*score_tst[1]))
print('')
```

### 18.2.2 Model Developement (Model_2) : Changing Units, Dropout, Epoch and Batch_size

### 18.2.3 Model Developement (Model_3) : Changing Activation, Loss and Optimizer

```python
## Step- 2 : Creating First model
print('------------------ MODEL_3 ------------------')

# Creating an empty sequential model
model_3 = Sequential()

# Adding an input layer to the model i.e Layer 1 [unit : 10 to 1000 to 100] [act Relu to softmax]
model_3.add(Dense(100, input_dim = input_dim, activation = 'softmax', kernel_initializer = 'uniform'))

# Adding a hidden layer  [drop out : 0 to 0.1]
model_3.add(Dropout(0.1))

# Adding an output layer to the model  [softmax to sigmoid]
model_3.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

## Step- 3 : Compile the model [opt: SGD to RMSprop ]
model_3.compile(loss = 'binary_crossentropy', optimizer = 'RMSprop', metrics = ['accuracy'])

## Step- 4 : Fit the model [epochs: 5 to 500 to 50, batch_size: 1000 to 2000 to 2000]
model_3.fit(x_train, y_train, epochs = 50, batch_size = 2000)
```

```python
## Step- 5 : Evaluate the model
score_trn = model_3.evaluate(x_train, y_train)
print('Training Accuracy is : %.2f%%' %(100*score_trn[1]))
print('')

score_tst = model_3.evaluate(x_test, y_test)
print('Test Accuracy is : %.2f%%' %(100*score_tst[1]))
print('')
```

#### Obs : Excellent Result. Accuracy increased to 99.83% for both train and test which means there is no overfitting and underfitting

### 18.2.4 Model Developement (Model_4) : Changing Activation & Optimizer

```python
## Step- 2 : Creating First model
print('------------------ MODEL_4 ------------------')

# Creating an empty sequential model
model_4 = Sequential()

# Adding an input layer to the model i.e Layer 1 [unit : 10 to 1000 to 100] [act Relu to softmax to relu]
model_4.add(Dense(1000, input_dim = input_dim, activation = 'relu', kernel_initializer = 'uniform'))

# Adding a hidden layer  [drop out : 0 to 0.1]
model_4.add(Dropout(0.1))

# Adding an output layer to the model  [softmax to sigmoid]
model_4.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

## Step- 3 : Compile the model [opt: SGD to RMSprop to adagrad ]
model_4.compile(loss = 'binary_crossentropy', optimizer = 'adagrad', metrics = ['accuracy'])

## Step- 4 : Fit the model [epochs: 5 to 500 to 50, batch_size: 1000 to 2000 to 2000]
model_4.fit(x_train, y_train, epochs = 50, batch_size = 2000)
```

### 18.2.5 Hyper parameter - Grid Approach to Determine Best value of Epoch and Batch Size
- To improve the accuracy we need to determine the best value of epoch and best size

```python
## Step- 2 : Creating Fifth model by using Grid search
from keras.constraints import maxnorm
print('------------------ MODEL_5 ------------------')

# Values of weight_constraints can be used are : [0,1,2,3,4,5]
weight_constraint = 0

# Creating a function
def create_model():
    model = Sequential()
    
    # Using maxnorm constraints
    model.add(Dense(1000, input_dim = input_dim, activation = 'relu', kernel_initializer = 'uniform',
                   kernel_constraint = maxnorm(weight_constraint)))
    model.add(Dropout(0.1))
    model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    # Compile model
    model.compile(loss = 'binary_crossentropy', optimizer = 'adagrad', metrics = ['accuracy'])
    
    # Return the model
    return model

## Create model using the above defined function
from keras.wrappers.scikit_learn import KerasClassifier
model_5 = KerasClassifier(build_fn = create_model)

## Creating a Grid of different Epochs and batch_size
from sklearn.model_selection import GridSearchCV
epochs = [50,100]
batch_size = [1500,2500,3000]

## Creating a dictionary
pram_grid = dict(epochs = epochs, batch_size = batch_size)

## Performing Grid search w.r.t epochs and batch_size
grid = GridSearchCV(estimator = model_5, param_grid = pram_grid)
grid_result = grid.fit(x_train, y_train)

## Display the complete results
print('The Complete Results: \n',grid_result.cv_results_)
print('')

# Display the summarize results
print('Best: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
```

```python
## Evaluating the model for the train and test dataset with best parameters
model_5.fit(x_train, y_train, epochs = grid_result.best_params_['epochs'], batch_size = grid_result.best_params_['batch_size'])

print('Training Accuracy of the new model after tuning the parameters : %.2f%%' %(100*model_5.score(x_train, y_train)))
print('')

model_5.fit(x_test, y_test, epochs = grid_result.best_params_['epochs'], batch_size = grid_result.best_params_['batch_size'])

print('Test Accuracy of the new model after tuning the parameters : %.2f%%' %(100*model_5.score(x_test, y_test)))
print('')
```

## Obs : The result shows that the model is neither overfitting nor underfitting as both train and test accuracy is 99.83% which is preety good
